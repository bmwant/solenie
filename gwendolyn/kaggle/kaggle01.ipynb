```python
import math
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.svm import NuSVC
```

```python
data = pd.read_csv('train.csv')
```

```python
def get_train_and_validation_data(data, ratio=0.8):
    len_train = math.floor(len(data)*ratio)
    train_data = data[:len_train]
    validation_data = data[len_train:]
    return train_data, validation_data
```

```python
train_data, validation_data = get_train_and_validation_data(data)
```

```python
train_data
```

```python
train_data['Age'].isnull().values.any()
```

```python
train_data['Age'].fillna(0, inplace=True)
# validation_data = validation_data['Age'].fillna(0)
```

```python
validation_data['Age'].fillna(0, inplace=True)
```

```python
validation_data
```

```python
classifier = NuSVC(gamma='scale', verbose=True)
```

```python
def get_X(data):
    arrays = []
    for index, row in data.iterrows():
        arrays.append(np.array(get_features(row)))
    return np.vstack(arrays)
```

```python
def get_y(data):
    array = []
    for index, row in data.iterrows():
        array.append(row['Survived'])
    return np.array(array)
```

```python
def get_features(item):
    return [item['Pclass'], item['Age'], item['SibSp'], item['Parch'], item['Fare']]
```

```python
X_train = get_X(train_data)
X_validate = get_X(validation_data)
```

```python
y_train = get_y(train_data)
y_validate = get_y(validation_data)
```

```python
classifier.fit(X_train, y_train)
```

```python
# classifier.score
# classifier.predict
```

```python
classifier.score(X_validate, y_validate)
```

```python
test_data = pd.read_csv('test.csv')
```

```python
test_data['Fare'].fillna(0, inplace=True)
```

```python
X_test = get_X(test_data)
```

```python
X_test
```

```python
test_prediction = classifier.predict(X_test)
```

```python
test_prediction
```

```python
def get_prediction_dataframe(test_data, prediction):
    df = pd.DataFrame(columns=['PassengerId', 'Survived'])
    for index, row in test_data.iterrows():
        df.loc[index] = [row['PassengerId'], prediction[index]]
    return df
```

```python
prediction_df = get_prediction_dataframe(test_data, test_prediction)
```

```python
def write_prediction(prediction_df):
    prediction_df.to_csv('prediction.csv', header=True, index=False)
```

```python
write_prediction(prediction_df)  # kaggle says 0.61244
```
